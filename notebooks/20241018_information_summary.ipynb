```python
import os
from dotenv import load_dotenv
load_dotenv()

import requests
from bs4 import BeautifulSoup

import re
```

```python
def get_news_text(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    text = soup.get_text()
    # 空白文字を1つの空白に圧縮
    text = re.sub(r' +', ' ', text)
    # 連続する改行を1つの改行に圧縮
    text = re.sub(r'\n+', '\n', text)
    # 連続するタブを1つのタブに圧縮
    text = re.sub(r'\t+', '\t', text)
    return text
```

```python
news_info = []
```

```python


urls = ["https://corporate.jp.sharp/news/"]
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    source_list = soup.find_all(class_="news_item_box clearfix")
    news_info.extend([
        ['シャープ',s.find(class_='news_item_date').text,s.find(class_ = 'news_item_tx').text,"https://corporate.jp.sharp/"+s.find('a').get('href')] 
        if s.find('a').get('href').startswith("/") else 
        ['シャープ',s.find(class_='news_item_date').text,s.find(class_ = 'news_item_tx').text,s.find('a').get('href')]
        for s in source_list 
        ])

```

```python

urls = ["https://www.konicaminolta.com/jp-ja/newsroom/index.html"]
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    source_list = soup.find_all(class_="newsroom-release-list")
    news_info.extend([
        ['コニカミノルタ',s.find(class_='newsroom-release-list__txt__date').contents[0].text,s.find_all('a', id=lambda x: x and x.startswith('newsroom_text'))[0].text,"https://www.konicaminolta.com"+s.find('a').get('href')] 
        if s.find('a').get('href').startswith("/") else 
        ['コニカミノルタ',s.find(class_='newsroom-release-list__txt__date').contents[0].text,s.find_all('a', id=lambda x: x and x.startswith('newsroom_text'))[0].text,s.find('a').get('href')] 
        for s in source_list 
        ])

```

```python

urls = ["https://jp.ricoh.com/release/2024","https://jp.ricoh.com/info/2024"]
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    source_list = soup.find_all("a", class_="news_lst")
    news_info.extend([
        ['リコー',s.find(class_='news_lst_day w160').text,s.find(class_='txt').text,"https://jp.ricoh.com"+s.get('href')] 
        if s.get('href').startswith("/") else 
        ['リコー',s.find(class_='news_lst_day w160').text,s.find(class_='txt').text,s.get('href')] 
        for s in source_list 
        ])
```

```python


urls = ["https://www.fujifilm.com/jp/ja/news/list/1/all"]
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    source_list = soup.find_all(class_="c-news-list__list")[0].find_all(class_="btn -link")
    date_list = soup.find_all(class_="c-news-list__list")[0].find_all('dt')
    news_info.extend([
        ['富士フイルム',t.text,s.text,s.get('href')] 
        for s,t in zip(source_list,date_list) 
        ])

```
