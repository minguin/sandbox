# LangChainのExampleを動かす

```python
from langchain.agents import load_tools
from langchain.agents import initialize_agent
from langchain.llms import OpenAI
from langchain.agents.mrkl import prompt
```

```python
# https://platform.openai.com/account/billing/overview
import os
os.environ["OPENAI_API_KEY"] = ""
os.environ["SERPAPI_API_KEY"] = ""
```

```python
# llm = OpenAI(temperature=0)
llm = OpenAI(temperature=0, model_name="gpt-3.5-turbo")
tools = load_tools(["serpapi", "llm-math"], llm=llm)
agent = initialize_agent(tools, llm, agent="zero-shot-react-description", verbose=True,
                         agent_kwargs=dict(suffix='Answer should be in Japanese.' + prompt.SUFFIX))
```

```python
agent.run("ブラックペレット技術を持つ、代表的なスタートアップ企業を10社程度、日本語で教えてください。")
```

```python
agent.run(input="ブラックペレット関連の重要な特許を持つ、代表的な企業を日本語で教えてください。")
```

```python
agent.run("脱炭素領域技術の最新トピック、技術名、キーワードを日本語で教えてください。")
```
